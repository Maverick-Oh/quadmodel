This cell loads the forward-modeled fluxes and corresponding parameters for ten mock lenses. To get the fluxes and parameters, you'll need to download them from google drives: https://drive.google.com/drive/folders/1jveqqGK9nh_h8PTLm7wlMNCPkIc4LnGe?usp=sharing
and place them in this directory

```python
def flux_ratios_from_fluxes(f, flux_uncertainties=[0.0001,0.0001,0.0001,0.0001]):
    
    delta_f = np.zeros_like(f)
    fluxes = np.empty_like(f)
    for i in range(0,4):
        delta_f[:,i] = f[:,i] * flux_uncertainties[i]
    fluxes = f + delta_f
    flux_ratios = fluxes[:,1:]/fluxes[:,0,None]
    return flux_ratios

# If you're downloading this from github, leave this argument as False
# You'll need to download the fluxes and parameters file from google drives: 
# https://drive.google.com/drive/folders/1jveqqGK9nh_h8PTLm7wlMNCPkIc4LnGe?usp=sharing
LOAD_RAW_SAMPLES = False

if LOAD_RAW_SAMPLES:
    nfiles = 2000
    fnames = ['mock1_fixedshear', 'mock2_fixedshear', 'mock3_fixedshear', 'mock4_fixedshear','mock5_fixedshear','mock6_fixedshear','mock7_fixedshear',
             'mock8_fixedshear','mock9_fixedshear','mock10_fixedshear']

    for fname in fnames:
        for i in range(0, nfiles):
            if i==0:
                parameters = np.loadtxt(fname+'/job_'+str(1)+'/parameters.txt',skiprows=1)
                fluxes = np.loadtxt(fname+'/job_'+str(1)+'/fluxes.txt')
            else:
                try:
                    parameters = np.vstack((parameters, np.loadtxt(fname+'/job_'+str(i+1)+'/parameters.txt',skiprows=1)))
                    fluxes = np.vstack((fluxes, np.loadtxt(fname+'/job_'+str(i+1)+'/fluxes.txt')))
                except:
                    continue

        np.savetxt(fname[:-11]+'_fluxes.txt',X=fluxes)
        np.savetxt(fname[:-11]+'_parameters.txt',X=parameters)
        
fluxes_1 = np.loadtxt('mock1_fluxes.txt')
parameters_1 = np.loadtxt('mock1_parameters.txt')

fluxes_2 = np.loadtxt('mock2_fluxes.txt')
parameters_2 = np.loadtxt('mock2_parameters.txt')

fluxes_3 = np.loadtxt('mock3_fluxes.txt')
parameters_3 = np.loadtxt('mock3_parameters.txt')

fluxes_4 = np.loadtxt('mock4_fluxes.txt')
parameters_4 = np.loadtxt('mock4_parameters.txt') 

fluxes_5 = np.loadtxt('mock5_fluxes.txt')
parameters_5 = np.loadtxt('mock5_parameters.txt')

fluxes_6 = np.loadtxt('mock6_fluxes.txt')
parameters_6 = np.loadtxt('mock6_parameters.txt')

fluxes_7 = np.loadtxt('mock7_fluxes.txt')
parameters_7 = np.loadtxt('mock7_parameters.txt')

fluxes_8 = np.loadtxt('mock8_fluxes.txt')
parameters_8 = np.loadtxt('mock8_parameters.txt')

fluxes_9 = np.loadtxt('mock9_fluxes.txt')
parameters_9 = np.loadtxt('mock9_parameters.txt')

fluxes_10 = np.loadtxt('mock10_fluxes.txt')
parameters_10 = np.loadtxt('mock10_parameters.txt')

flux_ratios_1 = flux_ratios_from_fluxes(fluxes_1)
flux_ratios_2 = flux_ratios_from_fluxes(fluxes_2)
flux_ratios_3 = flux_ratios_from_fluxes(fluxes_3)
flux_ratios_4 = flux_ratios_from_fluxes(fluxes_4)
flux_ratios_5 = flux_ratios_from_fluxes(fluxes_5)
flux_ratios_6 = flux_ratios_from_fluxes(fluxes_6)
flux_ratios_7 = flux_ratios_from_fluxes(fluxes_7)
flux_ratios_8 = flux_ratios_from_fluxes(fluxes_8)
flux_ratios_9 = flux_ratios_from_fluxes(fluxes_9)
flux_ratios_10 = flux_ratios_from_fluxes(fluxes_10)

mock1 = Mock1()
mock2 = Mock2()
mock3 = Mock3()
mock4 = Mock4()
mock5 = Mock5()
mock6 = Mock6()
mock7 = Mock7()
mock8 = Mock8()
mock9 = Mock9()
mock10 = Mock10()

fluxes_data_1 = mock1.m
flux_ratios_data_1 = fluxes_data_1[1:]/fluxes_data_1[0]

fluxes_data_2 = mock2.m
flux_ratios_data_2 = fluxes_data_2[1:]/fluxes_data_2[0]

fluxes_data_3 = mock3.m
flux_ratios_data_3 = fluxes_data_3[1:]/fluxes_data_3[0]

fluxes_data_4 = mock4.m
flux_ratios_data_4 = fluxes_data_4[1:]/fluxes_data_4[0]

fluxes_data_5 = mock5.m
flux_ratios_data_5 = fluxes_data_5[1:]/fluxes_data_5[0]

fluxes_data_6 = mock6.m
flux_ratios_data_6 = fluxes_data_6[1:]/fluxes_data_6[0]

fluxes_data_7 = mock7.m
flux_ratios_data_7 = fluxes_data_7[1:]/fluxes_data_7[0]

fluxes_data_8 = mock8.m
flux_ratios_data_8 = fluxes_data_8[1:]/fluxes_data_8[0]

fluxes_data_9 = mock9.m
flux_ratios_data_9 = fluxes_data_9[1:]/fluxes_data_9[0]

fluxes_data_10 = mock10.m
flux_ratios_data_10 = fluxes_data_10[1:]/fluxes_data_10[0]
```

### Likelihood-free inference

We now select sets of parameters that fix all three flux ratios simultaneously. First, we compute the summary statistic between the observed and model-predicted flux ratios

```python
def summary_stat(fdata,fmodel):
    s = []
    for i in range(0,fmodel.shape[0]):
        df = fmodel[i,:]-fdata
        s.append(np.sqrt(np.sum(df**2)))
    return np.array(s)

s1 = summary_stat(flux_ratios_data_1, flux_ratios_1)
s2 = summary_stat(flux_ratios_data_2, flux_ratios_2)
s3 = summary_stat(flux_ratios_data_3, flux_ratios_3)
s4 = summary_stat(flux_ratios_data_4, flux_ratios_4)
s5 = summary_stat(flux_ratios_data_5, flux_ratios_5)
s6 = summary_stat(flux_ratios_data_6, flux_ratios_6)
s7 = summary_stat(flux_ratios_data_7, flux_ratios_7)
s8 = summary_stat(flux_ratios_data_8, flux_ratios_8)
s9 = summary_stat(flux_ratios_data_9, flux_ratios_9)
s10 = summary_stat(flux_ratios_data_10, flux_ratios_10)
```

Next, we select the 2500 samples for each lens that correspond to the lowest summary statistics. The distribution of the statistics themselves are shown in the histogram. For reference, $S = 0.05$ correponds to matching each flux ratio to $\frac{S}{\sqrt{3}}=2.8\%$ precision

```python
# now we will keep the top "Nkeep" samples with the lowest summary statistics. 
# We can make a histogram of the accepted statistics to check that they fit the data well

Nkeep = 2500
inds1 = np.argsort(s1)[0:Nkeep]
inds2 = np.argsort(s2)[0:Nkeep]
inds3 = np.argsort(s3)[0:Nkeep]
inds4 = np.argsort(s4)[0:Nkeep]
inds5 = np.argsort(s5)[0:Nkeep]
inds6 = np.argsort(s6)[0:Nkeep]
inds7 = np.argsort(s7)[0:Nkeep]
inds8 = np.argsort(s8)[0:Nkeep]
inds9 = np.argsort(s9)[0:Nkeep]
inds10 = np.argsort(s10)[0:Nkeep]

fig = plt.figure(1)
fig.set_size_inches(9,7)
for i, (inds, stat) in enumerate(zip([inds1, inds2, inds3, inds4, inds5, inds6, inds7, inds8, inds9, inds10], (s1, s2, s3, s4, s5, s6, s7, s8, s9, s10))):
    plt.hist(stat[inds],alpha=0.5,label='mock '+str(i+1))
plt.legend(loc=2, frameon=False,fontsize=14)
plt.xlabel('S', fontsize=18)

samples_1 = parameters_1[inds1,0:2]
samples_2 = parameters_2[inds2,0:2]
samples_3 = parameters_3[inds3,0:2]
samples_4 = parameters_4[inds4,0:2]
samples_5 = parameters_5[inds5,0:2]
samples_6 = parameters_6[inds6,0:2]
samples_7 = parameters_7[inds7,0:2]
samples_8 = parameters_8[inds8,0:2]
samples_9 = parameters_9[inds9,0:2]
samples_10 = parameters_10[inds10,0:2]
```

### Now we can do the inference. The overall number of halos ($\Sigma_{\rm{sub}}$) is correlated with the warmth of dark matter (mhm, or $\log_{10}$ half-mode mass)

```python
from trikde.triangleplot import TrianglePlot
from trikde.pdfs import IndepdendentLikelihoods, DensitySamples, CustomPriorHyperCube

param_names = ['sigma_sub','mhm']
param_ranges = [[0.0, 0.2], [4, 10]]
weights = None
nbins = 20
use_kde = True
density1 = DensitySamples(samples_1, param_names, weights, param_ranges, nbins=nbins, use_kde=use_kde)
density2 = DensitySamples(samples_2, param_names, weights, param_ranges, nbins=nbins, use_kde=use_kde)
density3 = DensitySamples(samples_3, param_names, weights, param_ranges, nbins=nbins, use_kde=use_kde)
density4 = DensitySamples(samples_4, param_names, weights, param_ranges, nbins=nbins, use_kde=use_kde)
density5 = DensitySamples(samples_5, param_names, weights, param_ranges, nbins=nbins, use_kde=use_kde)
density6 = DensitySamples(samples_6, param_names, weights, param_ranges, nbins=nbins, use_kde=use_kde)
density7 = DensitySamples(samples_7, param_names, weights, param_ranges, nbins=nbins, use_kde=use_kde)
density8 = DensitySamples(samples_8, param_names, weights, param_ranges, nbins=nbins, use_kde=use_kde)
density9 = DensitySamples(samples_9, param_names, weights, param_ranges, nbins=nbins, use_kde=use_kde)
density10 = DensitySamples(samples_10, param_names, weights, param_ranges, nbins=nbins, use_kde=use_kde)


like = IndepdendentLikelihoods([density1,density2,density3,density4,density5,density6,density7,density8,density9,density10])

triplot = TrianglePlot([like])
triplot.make_triplot(filled_contours=False)
```
